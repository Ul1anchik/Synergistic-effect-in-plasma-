```python
# ток увлечения

def subtract(a, b):
    """
    Вычитает меньшее по модулю число из большего по модулю
    """
    if abs(a) >= abs(b):
        return abs(a) - abs(b)
    else:
        return abs(b) - abs(a)
    
a_left = 0.01889835585902
a_right = -0.02283317891505
print(subtract(a_left, a_right))    

# вклад в пик 

peak_left = 1.92096139690503
peak_right = -1.00954144805046
percent_left = a_left / peak_left * 100
percent_right = a_right / peak_right * 100

print(f"Процент: {percent_left}% в левое крыло")
print(f"Процент: {percent_right}% в правое крыло")

```
