# Импорт необходимых библиотек

Импортируйте numpy, matplotlib.pyplot и mpl_toolkits.mplot3d для визуализации.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

# Загрузка данных из .dat файла

Используйте numpy.loadtxt или pandas.read_csv для чтения данных из .dat файла с тремя столбцами: Ntor, Npol, Px.

```python
# Замените 'your_file.dat' на путь к вашему .dat файлу
data = np.loadtxt('your_file.dat')
Ntor = data[:, 0]
Npol = data[:, 1]
Px = data[:, 2]
```

# Преобразование данных в сетку для визуализации

Преобразуйте данные в двумерную сетку с помощью numpy.meshgrid для построения 2D и 3D графиков.

```python
# Получаем уникальные значения для сетки
ntor_unique = np.unique(Ntor)
npol_unique = np.unique(Npol)

# Создаем сетку
Ntor_grid, Npol_grid = np.meshgrid(ntor_unique, npol_unique)

# Создаем двумерный массив значений Px
Px_grid = np.full(Ntor_grid.shape, np.nan)
for i, n_pol in enumerate(npol_unique):
    for j, n_tor in enumerate(ntor_unique):
        mask = (Ntor == n_tor) & (Npol == n_pol)
        if np.any(mask):
            Px_grid[i, j] = Px[mask][0]
```

# Построение 2D-графика (imshow)

Постройте 2D-график распределения Px по осям Ntor и Npol с помощью plt.imshow.

```python
plt.figure(figsize=(6, 6))
plt.imshow(Px_grid, origin='lower', aspect='auto',
           extent=[ntor_unique.min(), ntor_unique.max(), npol_unique.min(), npol_unique.max()],
           cmap='plasma')
plt.xlabel('Ntor')
plt.ylabel('Npol')
plt.title('2D распределение Px')
plt.colorbar(label='Px')
plt.show()
```

# Построение 3D-графика (scatter3D)

Постройте 3D scatter-график, где оси — Ntor, Npol, а цвет/ось Z — Px, используя Axes3D.scatter.

```python
fig = plt.figure(figsize=(7, 6))
ax = fig.add_subplot(111, projection='3d')

# Для scatter нужны одномерные массивы
mask = ~np.isnan(Px_grid)
ax.scatter(Ntor_grid[mask], Npol_grid[mask], Px_grid[mask], c=Px_grid[mask], cmap='plasma', marker='o')

ax.set_xlabel('Ntor')
ax.set_ylabel('Npol')
ax.set_zlabel('Px')
plt.title('3D scatter Px(Ntor, Npol)')
plt.show()
```
